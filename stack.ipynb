```python
## 함수





## 전역
SIZE = 5
stack = [None for _ in range(SIZE)]
top = -1


## 메인
# Push() : 삽입입
top += 1
stack[top] = '커피'
top += 1
stack[top] = '녹차'
top += 1
stack[top] = '꿀물'

print('바닥:',stack)
```

```python
# Pop() : 추출(=삭제)
data = stack[top]
stack[top] = None
top -= 1
print('pop-->', data)

data = stack[top]
stack[top] = None
top -= 1
print('pop-->', data)

data = stack[top]
stack[top] = None
top -= 1
print('pop-->', data)

print('바닥:',stack)
```
